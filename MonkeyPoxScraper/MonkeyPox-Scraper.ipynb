```python
import pandas as pd
import json
import datetime as dt
import numpy as np
```

```python
bbc = pd.read_json('bbc_article_data.json', encoding='utf8')
bbc['outlet'] = 'The BBC'
bbc = bbc.dropna(axis=0, subset=['text'])
bbc = bbc.loc[bbc["text"] != "" ]
bbc['date'] = pd.to_datetime(bbc.date).dt.tz_localize(None)
bbc
```

```python
guardian = pd.read_json('guardian_article_data.json', encoding='utf8')
guardian['outlet'] = 'The Guardian'
guardian = guardian.dropna(axis=0, subset=['text'])
guardian = guardian.loc[guardian["text"] != "" ]
guardian['date'] = guardian['date'].str.replace('BST','')
guardian['date'] = guardian['date'].str.replace('GMT','')
for index, row in guardian.iterrows():
    if row['date'] == None:
        continue
    else:
        row['date'] = "".join(str(row['date']).rstrip().lstrip())
        d = dt.datetime.strptime(row['date'], "%a %d %b %Y %H.%M")
        row['date'] = d
guardian
```

```python
mail = pd.read_json('mail_article_data.json', encoding='utf8')
mail['outlet'] = 'The Daily Mail'
mail = mail.dropna(axis=0, subset=['text'])
mail = mail.loc[mail["text"] != "" ]
mail
```

```python
metro = pd.read_json('metro_article_data.json', encoding='utf8')
metro['outlet'] = 'The Metro'
metro = metro.dropna(axis=0, subset=['text'])
metro = metro.loc[metro["text"] != "" ]
metro
```

```python
mirror = pd.read_json('mirror_article_data.json', encoding='utf8')
mirror['outlet'] = 'The Mirror'
mirror = mirror.dropna(axis=0, subset=['text'])
mirror = mirror.loc[mirror["text"] != "" ]
mirror
```

```python
standard = pd.read_json('standard_article_data.json', encoding='utf8')
standard['outlet'] = 'The Evening Standard'
standard = standard.dropna(axis=0, subset=['text'])
standard = standard.loc[standard["text"] != "" ]
standard
```

```python
sun = pd.read_json('sun_article_data.json', encoding='utf8')
sun['outlet'] = 'The Sun'
sun = sun.dropna(axis=0, subset=['text'])
sun = sun.loc[sun["text"] != "" ]
sun
```

```python
telegraph = pd.read_json('telegraph_article_data.json', encoding='utf8')
telegraph['outlet'] = 'The Telegraph'
telegraph = telegraph.dropna(axis=0, subset=['text'])
telegraph = telegraph.loc[telegraph["text"] != "" ]
telegraph['title'] = telegraph['title'].str.replace('\n','')
telegraph['date'] = telegraph['date'].str.replace('\n','')
telegraph['date'] = telegraph['date'].str.replace('•','')
for index, row in telegraph.iterrows():
    if row['date'] == None:
        continue
    else:
        row['date'] = "".join(str(row['date']).rstrip().lstrip())
        d = dt.datetime.strptime(row['date'], "%d %B %Y  %I:%M%p")
        row['date'] = d
for index, row in telegraph.iterrows():
    if row['author'] == []:
        row['author'] = None
    else:
        row['author'] = ' '.join(row['author'])
telegraph = telegraph.replace(r'^\s+$', np.nan, regex=True)
telegraph
```

```python
all_outlets = pd.concat([bbc, guardian, mail, metro, mirror, standard, sun], ignore_index=True)
```

```python
all_outlets.to_csv('MonkeyPox News Data.csv', index=False, encoding='utf8')
```
